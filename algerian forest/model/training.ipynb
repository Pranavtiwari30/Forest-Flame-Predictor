```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
%matplotlib inline
```

```python
dataset=pd.read_csv('Algerian_forest_fires_dataset_UPDATE.csv', header=1)
```

```python
dataset.head()
```

### missing values


```python
dataset.isnull().sum()
```

```python
dataset.isnull().any(axis=1)
```

```python
dataset[dataset.isnull().any(axis=1)]
```

```python
dataset.loc[:122, "Region"]=0
dataset.loc[122:, "Region"]=1
df=dataset
```

```python
df.info()
```

```python
df[['Region']]=df[['Region']].astype(int)
```

```python
df.info()
```

```python
df.isnull().sum()
```

### removing the null values


```python
df=df.dropna().reset_index(drop=True)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.iloc[[122]]
```

```python
df=df.drop(122).reset_index(drop=True)
```

```python
df.iloc[[122]]
```

```python
df.columns
```

### fix spaces in columns

```python
df.columns=df.columns.str.strip()
df.columns
```

```python
df.info()
```

### change required columns as int data data type

```python
df[['day', 'month', 'year', 'Temperature', 'RH', 'Ws']]=df[['day', 'month', 'year', 'Temperature', 'RH', 'Ws']].astype(int)
```

```python
df.info()
```

### changing other columns to float data type

```python
objects = [features for features in df.columns if df[features].dtypes == 'object']
```

```python
for i in objects:
    if i!='Classes':
        df[i]=df[i].astype(float)
```

```python
df.info()
```

```python
df.describe()
```

```python
df.head()
```

### lets save cleaned datset

```python
df.to_csv('Algerian_forest_fires_cleaned_dataset.csv', index=False)
```
